## Análise Inferencial



A partir da Análise Exploratória, decidiu-se verificar a relação dos profissionais da saúde disponíveis nos municípios com a quantidade de estabelecimentos de saúde ofertado pelo SUS, pois os estabelecimentos de saúde não apresentam uma correlação com quantidade de profissionais ou a estimativa populacional, essa informação é interessante de ser observada, pois deveria exitir uma correlação entre essas variáveis.

Para isso foi adicionado uma nova coluna com o total de estabelecimentos (posto de Saúde, centro de saúde / unidade básica de saúde, policlinica, hospital geral, hospital especializado, pronto socorro geral e pronto socorro especializado) que o SUS oferece, sendo analisado de uma maneira geral para entender melhor a estrutura de saúde das cidades.

```python
# Converta as colunas de string em valores numéricos
df['HOSP_GER'] = pd.to_numeric(df['HOSP_GER'], errors='coerce')
df['HOSP_ESP'] = pd.to_numeric(df['HOSP_ESP'], errors='coerce')
df['PRT_SOC'] = pd.to_numeric(df['PRT_SOC'], errors='coerce')
df['PRT_ESP'] = pd.to_numeric(df['PRT_ESP'], errors='coerce')

# Crie uma nova coluna com a soma das colunas existentes
df['QTD_ESTAB'] = df['POSTO'] + df['CS_UBS'] + df['POLICL'] + \
              df['HOSP_GER'] + df['HOSP_ESP'] + df['PRT_SOC'] + \
              df['PRT_ESP']
```

Foi realizado a conversão das colunas HOSP_GER, HOSP_ESP, PRT_SOC e PRT_ESP de string para valores numéricos. E foi criado uma coluna chamada QTD_ESTAB que é a soma de todos os tipos de estabelecimentos de saúde ofertados.

```python
# colocando os nomes dos municipios em maiusculo e retirando todos os caracteres especiais
df['MUN'] = df['MUN'].str.upper()
df['MUN'] = df['MUN'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

# retirando os valores numericos com codigos das cidades e mantendo somente os nomes
df['MUN'] = df['MUN'].str.extract('(\D+)', expand=False).str.strip()
```

Foi colocado os nomes dos municipios em maiusculo, retirado todos os caracteres especiais, retirado os valores numericos com codigos das cidades e se manteve somente os nomes. 

```python
df['QTD_ESP'] = df['PSIQ'] + df['PSICO'] + df['ODONTO'] + \
              df['NUTRI'] + df['FONO'] + df['FISIO'] + \
              df['MD_URO'] + df ['MD_GERI'] + df['MD_FAM'] + df['MD_PED'] + \
              df['MD_CLIN'] + df['MD_GIN_OBS'] + df['MD_CIR_G']
```

Foi criada uma coluna QTD_ESP no qual representa o total da soma das colunas dos profissionais (psiquiatra, psicólogo, odontologista, nutricionista, fonoaudiólogo, médico urologista, médico geriatra, médico da familia, médico pediatra, médico clínico, médico ginecologista-obstetra, médico cirurgião geral) da saúde.

Foi desconsiderado as profissões de enfermeiro e assistente social, pois, embora elas sejam da área da saúde, elas só podem fazer triagens, não existindo a possibilidade de conseguir fazer um diagnostico.

```python
df['TAXA_ESTAB']= df['QTD_ESTAB']/df['QTD_ESP']*100
```

Foi criada uma coluna TAXA_ESTAB, onde a quantidade de estabelecimentos (QTD_ESTAB) divido pela quantidade de profissionais da saúde (QTD_ESP) e multiplicado por 100. Esse calculo nos indica a quantidade de profissionais por estabelecimentos em porcentagem .

```python
df=df.fillna(0)
df['TAXA_ESTAB'] = pd.cut(df['TAXA_ESTAB'], bins=[0, 15, 100], labels=False)
df=df.fillna(0)
```

```python
y=df['TAXA_ESTAB'] 
x=df.drop(columns=['QTD_ESTAB', 'CD_MUN', 'MUN', 'QTD_ESP', 'TAXA_ESTAB'])
```

Foi definido o meu Y (variável resposta). Que irá mostrar a densidade de estabelecimentos por profissional em cada município.

```python
# separando os datasets para treino do modelo e para o teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Ocorreu a separação dos dados em treino e teste, sendo tamanho de teste de 20% dos dados totais e fixando a semente aleatória para 42. Também foi realizado o escalonamento da tabela, para todas as variáveis terem valores de média 0 e desvio padrão 1 Foi realizado o escalonamento da tabela, para todas as variáveis terem valores de média 0 e desvio padrão 1.

```python
kf = KFold(n_splits=7, shuffle=True, random_state=11)
```

Foi realizada a reamostragem dos dados por Cross Validation, sendo separada em 7 folds aleatórios. 

### Decision Trees

```python
pipe = Pipeline([('model', DecisionTreeClassifier())])

scores = []
i = 1
for train_set, test_set in kf.split(x):
    x_train, x_test = x.loc[train_set], x.loc[test_set]
    y_train, y_test = y[train_set], y[test_set]
    pipe.fit(x_train, y_train)
    sco = pipe.score(x_test, y_test)
    scores.append(sco)
    print("iteration ", i)
    i += 1
```

```python
print('Scores: ', np.array(scores))

print('Mean: ', np.array(scores).mean())

print('SD: ', np.array(scores).std())
```

A acurácia média obtida foi de 99.14% com um desvio padrão de 0.16%.

```python
clf = DecisionTreeClassifier(max_depth=5)

clf.fit(x_train,y_train)

y_pred_train = clf.predict(x_train)
y_pred = clf.predict(x_test)

print("Train Accuracy:",metrics.accuracy_score(y_train, y_pred_train))
print("Test Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

A acurácia de treinamento foi de 0,9985, o que significa que o modelo classificou corretamente 99,85% das amostras no conjunto de treinamento. Já a acurácia de teste foi de 0,9949, ou seja, o modelo classificou corretamente 99,49% das amostras no conjunto de teste.
